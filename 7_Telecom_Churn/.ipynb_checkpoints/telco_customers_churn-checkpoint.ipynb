# Telecom Customers Churn

Data source: https://www.kaggle.com/blastchar/telco-customer-churn

```python
import keras
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.metrics import roc_auc_score,roc_curve
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score,recall_score

from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout, BatchNormalization
```

```python
PLOTS = True
random_seed = 342
np.random.seed(random_seed)
tf.random.set_seed(random_seed)
```

## Data
Import data and print some characteristics. 

```python
tlc = pd.read_csv(r'~/tf-keras/7_Telecom_Churn/Telco-Customer-Churn.csv')
tlc = tlc.drop('customerID', axis=1)
```

```python
print('Number of columns:    ', tlc.shape[1])
print('Number of row:        ', tlc.shape[0])
print ("\nMissing values :  ", tlc.isnull().sum().values.sum())
print ("\nUnique values :  \n",tlc.nunique())
```

```python
tlc.dtypes
```

Perform some data cleaning

```python
# convert total charges to numeric
tlc = tlc[tlc.TotalCharges != " "]
tlc.TotalCharges = pd.to_numeric(tlc['TotalCharges'])

# replace 'no internet service' with 'No'
replace_cols = [ 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                'TechSupport','StreamingTV', 'StreamingMovies']
for i in replace_cols : 
    tlc[i]  = tlc[i].replace({'No internet service' : 'No'})
```

```python
# create a dummy variable for target
tlc['Churn_Dummy'] = pd.get_dummies(tlc['Churn'], drop_first=True)
tlc['Churn_Dummy'] = pd.to_numeric(tlc['Churn_Dummy'])
```

## Exploratory Data Analysis

```python
if PLOTS:
    sns.pairplot(tlc)
    plt.show()
```

```python
if PLOTS == True:
    sns.boxplot(tlc['MonthlyCharges'], tlc['Churn'])
    plt.show()
```

## Prepare data for machine learning

```python
bin_vars = []
cat_vars = []
num_vars = []
for i in tlc.columns:
    if tlc.nunique()[i] == 2:
        bin_vars.append(i)
    elif tlc.nunique()[i] < 10:
        cat_vars.append(i)
    else:
        num_vars.append(i)
```

Label encode binary variables:

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in bin_vars:
      tlc[i] = le.fit_transform(tlc[i])
      print(f'column {i} converted to binary (numerical)')
```

One-hot-encode multiclass categorical columns:

```python
tlc = pd.get_dummies(data = tlc, columns = cat_vars)
```

Standardize numerical variables:

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
scaled_tlc = std.fit_transform(tlc[num_vars])
scaled_tlc = pd.DataFrame(scaled_tlc,columns=num_vars)
```

Drop original (non-scaled) columns and join back together numerical and categorical columns.

```python
df_telcom_og = tlc.copy()
tlc = tlc.drop(columns = num_vars, axis = 1)
tlc = tlc.drop(columns = 'Churn_Dummy', axis =1)
tlc = tlc.merge(scaled_tlc, left_index=True,right_index=True,how = "left")
tlc = tlc.dropna()
tlc.info()
tlc.describe()
```

```python
if PLOTS:
    sns.heatmap(tlc.corr(), cbar=True, square=True)
    plt.show()
```

Train-test set split

```python
from sklearn.model_selection import train_test_split
target_col = 'Churn'
train,test = train_test_split(tlc,test_size = .25 ,random_state = random_seed, stratify = tlc[target_col])
    
# seperating dependent and independent variables
cols    = [i for i in tlc.columns if i not in target_col]
train_X = train[cols]
train_Y = train[target_col]
test_X  = test[cols]
test_Y  = test[target_col]
```

## Logistic Regression

```python
logit  = LogisticRegression(C=1.0, class_weight='balanced', dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)

logit.fit(train_X,train_Y)
predictions   = logit.predict(test_X)
probabilities = logit.predict_proba(test_X)
coefficients  = pd.DataFrame(logit.coef_.ravel())
print (logit)
```

```python
print ("\n Classification report : \n",classification_report(test_Y,predictions))
print ("Accuracy   Score : ",accuracy_score(test_Y,predictions))
#confusion matrix
conf_matrix = confusion_matrix(test_Y,predictions)
print(conf_matrix)
#roc_auc_score
model_roc_auc = roc_auc_score(test_Y,predictions) 
print ("AUC-ROC score : ",model_roc_auc,"\n")
```

```python
fpr,tpr,thresholds = roc_curve(test_Y,probabilities[:,1])

plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False positives rate')
plt.ylabel('True Positives rate')
plt.title('Logistic ROC curve')
plt.show()
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc   = RandomForestClassifier(n_estimators = 1000,
                                   criterion = "entropy",
                                   class_weight = 'balanced',
                                  )
rfc.fit(train_X,train_Y)

preds   = rfc.predict(test_X)
probs = rfc.predict_proba(test_X)
coeffs  = pd.DataFrame(rfc.feature_importances_)
```

```python
print ("\n Classification report : \n",classification_report(test_Y,preds))
print ("Accuracy   Score : ",accuracy_score(test_Y,preds))
#confusion matrix
conf_matrix = confusion_matrix(test_Y,preds)
print(conf_matrix)
#roc_auc_score
model_roc_auc = roc_auc_score(test_Y,preds) 
print ("AUC-ROC score : ",model_roc_auc,"\n")
```

```python
fpr,tpr,thresholds = roc_curve(test_Y,probs[:,1])

plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False positives rate')
plt.ylabel('True Positives rate')
plt.title('RF ROC curve')
plt.show()
```

## FC network

```python
train_X = train_X.to_numpy()
train_Y = train_Y.to_numpy()
test_X = test_X.to_numpy()
test_Y = test_Y.to_numpy()
train_X.shape
```

```python
earlystop_cb = keras.callbacks.EarlyStopping(patience=10, restore_best_weights= True)
```

```python
model = Sequential([
        Dense(25, activation='relu', input_shape = train_X.shape[1:]),
        Dropout(0.2),
        Dense(25, activation='relu'),
        Dropout(0.2),
        Dense(1, activation = 'sigmoid')
        ])
model.compile(loss = 'binary_crossentropy', optimizer = 'nadam', metrics=['accuracy'])
history = model.fit(train_X, train_Y, epochs=30, validation_data=(test_X, test_Y), callbacks=[earlystop_cb])
```

```python
model1 = Sequential([
        Dense(25, activation='tanh', input_shape = train_X.shape[1:]),
        Dropout(0.2),
        Dense(25, activation='tanh'),
        Dropout(0.2),
        Dense(1, activation = 'sigmoid')
        ])
model1.compile(loss = 'binary_crossentropy', optimizer = 'nadam', metrics=['accuracy'])
history1 = model1.fit(train_X, train_Y, epochs=50, validation_data=(test_X, test_Y), callbacks=[earlystop_cb])
```

```python
model3 = Sequential([
        BatchNormalization(),
        Dense(25, activation='elu', kernel_initializer = 'he_normal', input_shape = train_X.shape[1:]),
        Dropout(0.2),
        BatchNormalization(),
        Dense(25, activation='elu', kernel_initializer = 'he_normal'),
        Dropout(0.2),
        BatchNormalization(),
        Dense(1, activation = 'sigmoid')
        ])
model3.compile(loss = 'binary_crossentropy', optimizer = 'nadam', metrics=['accuracy'])
history3 = model3.fit(train_X, train_Y, epochs=100, validation_data=(test_X, test_Y), callbacks=[earlystop_cb])
```

```python
pd.DataFrame(history.history).plot(figsize = (6,4))
plt.grid(True)
plt.show()

pd.DataFrame(history1.history).plot(figsize = (6,4))
plt.grid(True)
plt.show()

pd.DataFrame(history3.history).plot(figsize = (6,4))
plt.grid(True)
plt.show()
```

```python
model4 = Sequential([
        Dense(25, activation='tanh', input_shape = train_X.shape[1:]),
        Dropout(0.2),
        Dense(25, activation='tanh'),
        Dropout(0.2),
        Dense(1, activation = 'sigmoid')
        ])
model4.compile(loss = 'binary_crossentropy', optimizer = 'nadam', metrics=['accuracy'])
history4 = model4.fit(train_X, train_Y, epochs=50, validation_data=(test_X, test_Y), callbacks=[earlystop_cb])
```

```python
_, val_acc_0 = model.evaluate(test_X, test_Y)
_, val_acc_1 = model1.evaluate(test_X, test_Y)
_, val_acc_3 = model3.evaluate(test_X, test_Y)
print(f'Test set accuracy: \n Model_0: {val_acc_0} \n Model_1: {val_acc_1} \n Model_3: {val_acc_3}')
```
